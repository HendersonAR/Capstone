```python
data2 = (data[['transfer', 'player', 'age', 'season', 'minutes', 'matches_played', 'matches_no_play',
                "raw_goals", 'raw_assists', 'raw_nonpenaltykick_goal', 'total_pass_attempts', 'Total_Cmp.',
                'Total_Passes_Leading_to_Shot', 'Cmp_Passes_18_yard_box', 'Progressive_Passes', 'Total_Pass_Distance_Towards_Opp_Goal',
                'Shot_Creating_Actions', 'def_action_to_shot', 'goal_creating_action', 'takeon_to_goal', 'prog_carry_distance', 'prog_carries',
                'Progressive_Passes_Received', 'TakeOn_Attempts', 'TakeOn_Success_Percentage', 'Tackles_Won', 'Tackles_Total',
                'Shot_Blocks', 'Pass_Blocks', 'Clearances', 'Aerial_Win_Percentage', 'Aerials_Won', 'Aerials_Lost', 'Position' ]])
```

```python
data2.head()
```

```python
data2['FW'] = data2['Position'].str.contains('FW').astype(int)
data2['MF'] = data2['Position'].str.contains('MF').astype(int)
data2['DF'] = data2['Position'].str.contains('DF').astype(int)
```

```python
data2.head()
```

```python
data2['play_proportion'] = data2['matches_played'] / (data2['matches_played'] + data2['matches_no_play']) 
```

```python
data2['prog_dist_per_pass'] = data2['Total_Pass_Distance_Towards_Opp_Goal'] / data2['Progressive_Passes']
```

```python
data2['prog_dist_per_carry'] = data2['prog_carry_distance'] / data2['prog_carries']
```

```python
data2['tackle_ratio'] = data2['Tackles_Won'] / data2['Tackles_Total']
```

```python
data2['aerials_total'] = data2['Aerials_Lost'] + data2['Aerials_Won']
```

```python
data2.head()
```

```python
data2.columns
```

```python
data3 = (data2[['transfer', 'player', 'age', 'season', 'minutes', 'matches_played','play_proportion',
                'raw_goals', 'raw_assists', 'raw_nonpenaltykick_goal', 'total_pass_attempts', 'Total_Cmp.',
                'Total_Passes_Leading_to_Shot', 'Cmp_Passes_18_yard_box', 'prog_dist_per_pass', 
                'Shot_Creating_Actions', 'def_action_to_shot', 'goal_creating_action', 'takeon_to_goal',
                'prog_dist_per_carry', 'Progressive_Passes_Received', 'TakeOn_Attempts', 'TakeOn_Success_Percentage',
                'tackle_ratio', 'Shot_Blocks', 'Pass_Blocks', 'Clearances', 'Aerial_Win_Percentage', 'aerials_total',
                'FW', 'MF', 'DF']])
```

```python
data3['minutes'] = data3['minutes'].str.replace(",","").astype(float)
```

```python
data3.isna().sum()
```

```python
data3['tackle_ratio'] = data3['tackle_ratio'].fillna(0)
```

```python
data3.replace([np.inf, -np.inf], 0, inplace = True)
```
